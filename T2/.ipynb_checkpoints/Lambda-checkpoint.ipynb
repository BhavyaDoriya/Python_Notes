```python
y=lambda x:x**2
print(y(9))
```

```python
y=lambda x,z:x+z
print(y(3,4))
```

```python
# Map
L=[1,2,3,4]
print(list(map(lambda x:x**2,L)))
```

```python
L=eval(input("Enter list:"))
list(map(lambda x: "even" if x%2==0 else "odd",L)) 
```

```python
L=["VHA10","NAS9","DDP8","PDO9"]
list(map(lambda x:x.lower(),L))
```

```python
# filter
L=[1,2,3,4,5,6,7]
list(filter(lambda x:x%2!=0,L))
```

```python
L=["apple","ant","are","aaaa","ball","ballon","bike"]
list(filter(lambda x:x.startswith("a") and len(x)>3,L))
```

```python
# reduce
L=[1,2,3]
from functools import reduce
reduce(lambda x,y:x+y,L)
```
