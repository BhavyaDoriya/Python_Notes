```python
# sorted does not change original list
L=[["python",10],["FSD",9],["DE",9],["PS",8]]
def f(X):
    return X[1]
#will sort by marks 
print(sorted(L,key=f))
print(sorted(L,key=f,reverse=True))

def f1(X):
    return len(X[0])
# will sort by subject letter length
print(sorted(L,key=f1))

# L=[58858,5959,1,2,3,4,5]
# L.sort() -->WIll change original list,will only work in 1-D list
# print(L)
```
