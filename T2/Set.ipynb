```python
s1={1,2,3,4}
s2={3,4,5,6}
print(s1.union(s2))
print(s1 | s2)
print(s1.intersection(s2))
print(s1 & s2)
print(s1.difference(s2))
print(s1-s2)
print(s1.symmetric_difference(s2))
print(s1^s2)
```

```python
s3={1,2}
s4={1 ,2 ,3}
print(s3.issubset(s4))
print(s4.issubset(s3))
print(s4.issuperset(s3))
print(s3.update(s4))#same as union but changes original set
print(s3)
```
