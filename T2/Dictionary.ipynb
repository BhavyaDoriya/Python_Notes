```python
d={"I":1,"V":5,"X":10,"L":50,"C":100,"D":500,"M":1000}
s=input("Enter roman number:")
ans=0
for i in range(0,len(s)):
    if(i!=len(s)-1 and d[s[i]]<d[s[i+1]]):
        ans-=d[s[i]]
    else:
        ans+=d[s[i]]
print(ans)
```
