```python
class A:
    def __init__(self):
        print("Constructor A")
    def show(self):
        print("Show A")
    def xyz(self):
        print("XYZ A")
class B(A):
    def __init__(self):
        print("Constructor B")
        super().__init__()#use of super keyword
    def show(self):
        super().xyz()
        print("Show B")
        self.xyz()
# when there is no constructor in child class but there is constructor present in parebt class only then it calls parent's constructor
# No error if parent doesn't have no parameter constructor and child has it and vise versa
b=B()
b.show()#will call child's show method
```

```python
class A:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def show(self):
        print("A:",self.a,"B:",self.b,"C:",self.c)
    
class B(A):
    pass
b=B()
# will give error
# reason: No child constructor hence it will go to parent's class and check for no parameter constructor but
# there is no non-parameterized constructor hence it will ask for parameters
```

```python
class A:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
        self.show()
    
    def show(self):
        print("A:",self.a,"B:",self.b,"C:",self.c)
    
class B(A):
    def __init__(self,a,b,c,d):
        super().__init__(a,b,c)
        print("D:",d)
b=B(3,4,5,6)
```

```python
# Multilevel
class A:
    def __init__(self):
        print("A")
    def show(self):
        print("SHOW A")
class B(A):
    def __init__(self):
        print("B")
    def show(self):
        print("SHOW B")
class C(B):
    def __init__(self):
        print("C")
    def show(self):
        print("SHOW C")
c=C()
b=B()
a=A()
b.show()
c.show()
a.show()
```

```python
# Multiple
class Mitch:
    def __init__(self):
        print("Mitch constructor")
    def show(self):
        print("SHOW MITCH")
class Cam:
    def __init__(self):
        print("Cam constructor")
    def show(self):
        print("SHOW CAM")
class Lily(Cam,Mitch):
    def __init__(self):
        print("Lily Constructor") #if it is commented then it will call constructor of Cam class
    def show(self):
        print("SHOW Lily")
c=Lily()
b=Cam()
a=Mitch()
b.show()
c.show()
a.show()
```

```python
class A:
    def show(self):
        print("A")
class B:
    def show(self):
        print("B")
class C:
    def show(self):
        print("C")
class X(A,B):
    def show(self):
        print("X")
class Y(B,C):
    def show(self):
        print("Y")
class P(X,Y,C):
    pass
    
P.__mro__
print(P.mro())
p=P()
p.show()
```

```python
# Operator overloading
# Available operators:
# __add__
# __sub__
# __mul__
# __truediv__
# __floordiv__
# __mod__
# __pow__
# __eq__
# __ne__
# __gt__
# __lt__
# __le__
# __ge__
# __str__  -->same as toString in java
class A:
    def __init__(self,a):
        self.m=a
    def __add__(self,other):
        return self.m+other.m
    def __truediv__(self,other):
        return self.m/other.m
    def __floordiv__(self,other):
         return self.m//other.m
    def __str__(self):
        return str(self.m)
x=A(50)
y=A(23)
print(x)
print(x/y)
print(x//y)
print(x+y)
```
