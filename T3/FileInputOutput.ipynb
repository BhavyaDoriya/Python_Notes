```python
# a mode--> in a mode it appends the text to a file(does not overwrite it)
f=open("files\\1.txt","a")
f.writelines(["abc\n","def\n"])
# f.read()-->can only write
f.close()
```

```python
f=open("files/3.txt","r+")
print(f.write("VHA"))#will start overwriting the file from pointer 0
print(f.read())
f.close()
#r+ or r mode doesn't create new file if file does not exist
```

```python
f=open("files/3.txt","w+")
f.write("Welcome")#will delete all previous data and start writing from 0
print(f.read())#will not print anything cause pointer is at end of file
print(f.tell())#will tell current position of the pointer
f.seek(0)#will change the position of pointer\
# seek(offset,position)
# 0-->start , 1-->current,2-->end
print(f.read())
f.close()
```

```python
f=open("files/3.txt","a+")
print(f.read())#when you open with a+ mode your cursor moves to end of file by default so it will return empty string
f.close()
```

```python
f=open("files/1.txt")
for i in f:
    print(i,end="")
f.close()
```

```python
# to automatically close file
with open("files/1.txt") as f:
    for i in f:
        print(i,end="")
```

```python
with open("files/1.txt") as f:
    print(f.name)
    print(f.mode)
    print(f.closed)
    print(f.readable())
    print(f.writable())
```
