```python
# (1) and (2)
f=open("mbox-short.txt")
d={}
count=0
for i in f:
    if i.startswith("From "):
        print(i.split()[1])
        d[i.split()[1]]=d.get(i.split()[1],0)+1
        count+=1
print("Count=",count)
def f1(X):
    return X[1]
d=sorted(d.items(),key=f1,reverse=True)
print(d[0])
f.close()
```

```python
#(3)
f=open("mbox-short.txt")
d={}
count=0
for i in f:
    if i.startswith("From "):
        d[i.split()[5].split(":")[0]]=d.get(i.split()[5].split(":")[0],0)+1
        count+=1
print("Count=",count)
def f1(X):
    return X[1]
d=sorted(d.items(),key=f1,reverse=True)
print(d)
f.close()
```

```python
#(4)
f=open("mbox-short.txt")
s=0
count=0
for i in f:
    if i.startswith("X-DSPAM-Confidence: "):
        s+=float(i.split()[1])
        count+=1
print("Sum",s)
print("Avg",s/count)
f.close()
```

```python
#(5)
f=open("logfile.txt")
def calDiff(login,logout):
    loginMin=int(login.split(":")[1])/60
    logoutMin=int(logout.split(":")[1])/60
    loginHour=int(login.split(":")[0])
    logoutHour=int(logout.split(":")[0])
    diff=(logoutHour+logoutMin)-(loginHour+loginMin)
    return diff
for i in f:
    L=i.split()
    login=i[1][0:4]
    logout=i[2][0:4]
    x=calDiff(login,logout)
    print(x)
```
