```python
class Atm:
    def __init__(self,pin,balance):
        self.pin=pin
        self.balance=balance

obj1=Atm("3252",10)
print(obj1.pin)
print(obj1.balance)
print(id(obj1.balance))
print(id(obj1))

obj1.balance=10000
print(obj1.balance)
print(id(obj1.balance))
print(id(obj1))

```
