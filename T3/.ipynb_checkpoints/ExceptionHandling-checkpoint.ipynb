```python
try:
    bal=int(input("Enter balance: "))
    withdraw=int(input("Enter withdraw amount:"))
    if withdraw<0:
        raise Exception("Withdraw amount can't be negative")
    if bal<0:
        raise Exception("Balance amount can't be negative")
    if withdraw>bal:
        raise Exception("Insuffiecient balanace")
except Exception as e:
    print(e)
else:
    print("AMount withdrawn successfully")
```

```python
class Vishal(Exception):
    def __init__(self,message):
        pass
try:
    bal=int(input("Enter balance: "))
    withdraw=int(input("Enter withdraw amount:"))
    if withdraw<0:
        raise Exception("Withdraw amount can't be negative")
    if bal<0:
        raise Exception("Balance amount can't be negative")
    if withdraw>bal:
        raise Vishal("Insuffiecient balanace")
except Exception as e:
    print(e.with_traceback)
else:
    print("AMount withdrawn successfully")
```
