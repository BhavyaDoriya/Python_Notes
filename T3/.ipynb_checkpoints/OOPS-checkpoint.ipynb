```python
class Circle:
    def __init__(self,r):
        self.r=r
    def perimeter(self):
        return 3.14*2*self.r
    def area(self):
        return 3.14*self.r*self.r
    def __del__(self):
        print("Bye")
c=Circle(4)
print("Perimeter:",c.perimeter())
del c
print("Area:",c.area())
# when you don't delete an object and there is destructor present in your code 
# for the first run it doesn't call destructor
# but when you run for the second time it calls the destructor
```

```python
class Circle:
    def __init__(self,r):
        self.r=r
    def perimeter(self):
        return 3.14*2*self.r
    def area(self):
        return 3.14*self.r*self.r
    def __del__(self):
        print("Bye")
c=Circle(4)
print("Perimeter:",c.perimeter())
print("Area:",c.area())
```

```python

class Atm:
    def __init(self):
        self.bal=0
        self.pin=""
    def createPin(self):
        bal=int(input("Enter balance:"))
        if(bal<0):
            print("Balance can't be negative")
        else:
            pin=input("Enter your new pin:")
            self.bal=bal
            self.pin=pin
    def updatePin(self):
        verifyPin=input("Enter your old pin:")
        if(verifyPin==self.pin):
            newPin=input("Enter new pin:")
            self.pin=newPin
        else:
            print("The pin entered is incorrect")
    def withdraw(self):
        amt=int(input("Enter amount to withdraw:"))
#         min amount should be 1k
        if self.bal-amt<=1000:
            print("Insufficient balance")
        elif amt<0:
            print("amount can't be negative")
        else:
            print(amt,"amount withdrawed successfully")
            self.bal-=amt
    def menu(self):
        print("1. Create new pin")
        print("2. Update pin")
        print("3. Withdraw")
        print("4. Display balance")
        print("5. Exit")
    def display(self):
        print("Balance:",self.bal)
        
o=Atm()
while(True):
    o.menu()
    choice=int(input("Enter choice:"))
    if choice==1:
        o.createPin()
    elif choice==2:
        o.updatePin()
    elif choice==3:
        o.withdraw()
    elif choice==4:
        o.display()
    elif choice==5:
        break
    else:
        print("Please Enter valid choice")
```
